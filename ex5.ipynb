```python
list=[19,9,9,23,89,13,0]
```

```python
def avf(nums):
  sum=0
  for i in nums:
    sum=sum+i
  avg=sum/len(nums)
  return avg
```

```python
print(avf(list))
```

```python
def avf1(nums):
  return sum(nums)/len(nums)
print(avf1(list))
```
