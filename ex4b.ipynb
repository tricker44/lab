```python
from nltk.corpus import brown
```

```python
import nltk
```

```python
cf=nltk.ConditionalFreqDist((genre , word) for genre in brown.categories() for word in brown.words(categories=genre))
```

```python
cf.conditions()
```

```python
cf["hobbies"]
```

```python
cf.tabulate(conditions=["government","hobbies","humor"], samples=["leadership","worship","hardship"])
```

```python
cf.plot(conditions=["government","hobbies","humor"], samples=["leadership","worship","hardship"])
```
