```python
import nltk
import numpy as np
import pandas as pd
```

```python
iteams=['apple','apple','kiwi','kiwi','mango','banana']
fd=nltk.FreqDist(iteams)
```

```python
cf=[('f','apple'),('f','apple'),('k','banana')]
```

```python
cfd=nltk.ConditionalFreqDist(cf)
```

```python
cfd.conditions()
```

```python
cfd.plot()
```
