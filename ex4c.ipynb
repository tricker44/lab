```python
from nltk.corpus import names
```

```python
import nltk
```

```python
cf= [(fid.split('.')[0], name[-1])
     for fid in names.fileids()
     for name in names.words(fid)]
```

```python
cfd=nltk.ConditionalFreqDist(cf)
```

```python
cfd['male']>cfd['female']
```

```python
cfd.tabulate(samples=['a','e'])
```

```python
cfd.plot()
```
